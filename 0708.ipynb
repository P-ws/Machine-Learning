```python
from selenium import webdriver as wb
from bs4 import BeautifulSoup as bs
import time
import pandas as pd
```

```python
url = 'https://www.hsd.co.kr/menu/menu_list'
```

```python
driver.get(url)
```

```python
soup = bs(driver.page_source, 'html.parser')
```

```python
num = []
names = []
prices = []

name = soup.findAll('h4', class_='h fz_03')
price = soup.select('div.item-text > div > strong')
len(name), len(price)

for i in range(len(name)):
    num.append(i + 1)
    names.append(name[i].text)
    prices.append(price[i].text)

print(num)
print(names)
print(prices)
```

```python
info = {'num':num, 'name':name, 'price':price}
df = pd.DataFrame(info)
df.set_index('num',inplace=True)
df
```

```python
df.to_csv('lunchBox_info.csv', encoding='utf-8')
```

```python
from selenium import webdriver as wb
from bs4 import BeautifulSoup as bs
import time
import pandas as pd
```

```python
driver=wb.Chrome('연습/chromedriver.ipynb')
```

```python
url = 'http://www.istarbucks.co.kr/store/store_map.do'
driver.get(url)
```

```python
btn_local = driver.find_element_by_class_name('loca_search')
btn_local.click()
```

```python
# 광주지역 클릭
li = driver.find_element_by_css_selector('ul.sido_arae_box > li + li + li')
li.click()
time.sleep(1)
```

```python
all = driver.find_element_by_css_selector('ul.gugun_arae_box > li')
all.click()
time.sleep(1)
```

```python
soup = bs(driver.page_source, 'html.parser')
```

```python
names = []
addrs = []
tels = []

name = soup.select('#mCSB_3_container > ul > li > strong')
addr = soup.select('#mCSB_3_container > ul > li > p')

for i in range(len(name)):
  names.append( name[i].text )
  addrs.append( addr[i].text[:-9] )
  tels.append( addr[i].text[-9:] )
```

```python
info ={'name':names, 'addrs':addrs, 'tels':tels}
```

```python
a= pd.DataFrame(info)
```
